```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/preprocessed_data.csv")
df.head()
```

```python
# Plot 1: KDE Plot(Delay_Severity)
sns.kdeplot(
    data=df, 
    x='Delay_Severity', 
    hue='satisfaction_score', 
    fill=True, 
    common_norm=False, 
    palette='coolwarm', 
)
plt.title('Distribution of Delay_Severity by Satisfaction\n(KDE Plot)')
plt.xlabel('Delay_Severity')
plt.ylabel('Density')

plt.show()
```

```python
# Plot 1: KDE Plot(Soft_Service_Score)
sns.kdeplot(
    data=df, 
    x='Soft_Service_Score', 
    hue='satisfaction_score', 
    fill=True, 
    common_norm=False, 
    palette='coolwarm', 
)
plt.title('Distribution of Soft_Service_Score by Satisfaction\n(KDE Plot)')
plt.xlabel('Soft_Service_Score')
plt.ylabel('Density')

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Delay_Severity', y='Age_Group',
                alpha=0.5, edgecolor=None)
plt.title('Scatter Plot: Age_Group vs Delay Severity')
plt.xlabel('Delay Severity')
plt.ylabel('Age_Group')
plt.tight_layout()
plt.show()
```
