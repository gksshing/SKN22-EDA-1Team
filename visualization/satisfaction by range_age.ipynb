```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/preprocessed_data.csv")
```

```python
# 1) X축= Flight_Range, Y축 = satisfaction_score

plt.figure(figsize=(6,3))
sns.barplot(
    data=df,
    x='Flight_Range',
    y='satisfaction_score',
    estimator='mean',
    hue='Flight_Range',
    palette='Set1'
)

plt.title('Average Satisfaction Score by Flight Range')
plt.xlabel('Flight Range')
plt.ylabel('Average satisfaction_score')
plt.ylim(0, 1)
plt.show()

# 2) X축= Age_Group, Y축 = satisfaction_score

plt.figure(figsize=(6,3))
sns.barplot(
    data=df,
    x='Age_Group',
    y='satisfaction_score',
    estimator='mean',
    hue='Age_Group',
    palette='Set1'
)

plt.title('Average Satisfaction Score by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average satisfaction_score')
plt.ylim(0, 1)
plt.show()
```
