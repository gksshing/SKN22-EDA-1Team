```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/preprocessed_data.csv")
```

```python
sns.barplot(
    df,
    x = 'Hard_Service_Score',
    y = 'satisfaction_score',
    estimator= np.mean,
    hue='Hard_Service_Score',
    palette='summer'

)

plt.show()

```

```python
sns.barplot(
    df,
    x = 'Tech_Score',
    y = 'satisfaction_score',
    estimator= np.mean,
    hue='Tech_Score',
    palette='summer'

)

plt.show()

```
